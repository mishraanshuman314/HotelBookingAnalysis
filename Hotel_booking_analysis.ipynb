#### Chart - 1

```python
# Function to get count from a column for bar chart
def get_count_from_column_bar(df, column_label):
  df_grpd = df[column_label].value_counts()
  df_grpd = pd.DataFrame({'index':df_grpd.index, 'count':df_grpd.values})
  return df_grpd

# Function to plot a bar chart from a column
def plot_bar_chart_from_column(df, column_label, t1):
  df_grpd = get_count_from_column(df, column_label)
  fig, ax = plt.subplots(figsize=(14, 6))
  c= ['g','r','b','c','y']
  ax.bar(df_grpd['index'], df_grpd['count'], width = 0.4, align = 'edge', edgecolor = 'black', linewidth = 4, color = c, linestyle = ':', alpha = 0.5)
  plt.title(t1, bbox={'facecolor':'0.8', 'pad':3})
  plt.legend()
  plt.ylabel('Count')
  plt.xticks(rotation = 15) # use to format the lable of x-axis
  plt.xlabel(column_label)
  plt.show()
```

```python
# Function to get count from a column
def get_count_from_column(df, column_label):
  df_grpd = df[column_label].value_counts()
  df_grpd = pd.DataFrame({'index':df_grpd.index, 'count':df_grpd.values})
  return df_grpd

# Function to plot a pie chart from a column
def plot_pie_chart_from_column(df, column_label, t1, exp):
  df_grpd = get_count_from_column(df, column_label)
  fig, ax = plt.subplots(figsize=(14,9))
  ax.pie(df_grpd.loc[:, 'count'], labels=df_grpd.loc[:, 'index'], autopct='%1.2f%%',startangle=90,shadow=True, labeldistance = 1, explode = exp)
  plt.title(t1, bbox={'facecolor':'0.8', 'pad':3})
  ax.axis('equal')
  plt.legend()
  plt.show()
```

```python
# Define explode values for the pie chart
exp1 = [0.05,0.05]
# Plot a pie chart for the 'hotel' column with specified explode values
plot_pie_chart_from_column(hotel_booking_df, 'hotel', 'Booking percentage of Hotel by Name', exp1)
```

```python
# Function to get count from a column and return a DataFrame
def get_count_from_column(df, column_label):
    # Calculate the count of values in the specified column
    df_grpd = df[column_label].value_counts()
    # Create a DataFrame with 'index' and 'count' columns
    df_grpd = pd.DataFrame({'index': df_grpd.index, 'count': df_grpd.values})
    # Return the DataFrame
    return df_grpd

# Function to plot a bar chart from a column
def plot_bar_chart_from_column(df, column_label, title):
    # Get the count DataFrame from the specified column using the helper function
    df_grpd = get_count_from_column(df, column_label)
    # Create a bar chart with specified settings
    fig, ax = plt.subplots(figsize=(14, 6))
    c = ['g', 'r', 'b', 'c', 'y']
    # Plotting the bar chart with specified parameters
    ax.bar(df_grpd['index'], df_grpd['count'], width=0.4, align='edge', edgecolor='black', linewidth=4, color=c, linestyle=':', alpha=0.5)
    # Adding a title with a colored background for better visibility
    plt.title(title, bbox={'facecolor': '0.8', 'pad': 3})
    # Labeling the y-axis as 'Count'
    plt.ylabel('Count')
    # Rotating x-axis labels for better readability
    plt.xticks(rotation=15)
    # Labeling the x-axis with the specified column_label
    plt.xlabel(column_label)
    # Display the bar chart
    plt.show()
# Call the function to plot the bar chart for the percentage distribution of bookings made by agents
plot_bar_chart_from_column(hotel_booking_df, 'agent', 'Percentage Distribution of Bookings Made by Agent')
```

```python
def plot_pie_chart_from_column(df, column_label, title, explode):
    """
    Plots a pie chart from a column in a DataFrame.

    Args:
        df (pd.DataFrame): The DataFrame to plot from.
        column_label (str): The name of the column to plot.
        title (str): The title of the pie chart.
        explode (list): A list of the same length as the values in the `column_label` column,
                       specifying the amount by which each slice of the pie chart should be exploded.

    Returns:
        None
    """
    # Group the DataFrame based on the specified column
    df_grouped = df[column_label].value_counts()

    # Create subplots for the pie chart with a specified size
    fig, ax = plt.subplots(figsize=(10, 8))

    # Plot the pie chart with parameters such as labels, autopct (percentage format), startangle, explode, and shadow
    ax.pie(df_grouped, labels=df_grouped.index, autopct='%1.2f%%', startangle=90, explode=explode, shadow=True)

    # Add a title with a colored background for better visibility
    plt.title(title, bbox={'facecolor': '0.8', 'pad': 3})

    # Equal aspect ratio ensures that pie is drawn as a circle
    ax.axis('equal')

    # Display the pie chart
    plt.show()

# Assuming your dataframe is named hotel_booking_df
explode_deposit_type = (0.1, 0.1, 0.1)  # Explode the first slice (if needed)

# Example usage: Plotting the pie chart for deposit types
plot_pie_chart_from_column(hotel_booking_df, 'deposit_type', 'Percentage Distribution of Deposit Types', explode_deposit_type)
```

```python
def plot_bar_chart_from_column(df, column_label, title):
    """
    Plots a bar chart from a column in a DataFrame.

    Args:
        df (pd.DataFrame): The DataFrame to plot from.
        column_label (str): The name of the column to plot.
        title (str): The title of the bar chart.

    Returns:
        None
    """

    # Group the DataFrame based on the specified column and calculate the count for each group
    df_grouped = df[column_label].value_counts()

    # Create subplots for the bar chart with a specified size
    fig, ax = plt.subplots(figsize=(12, 8))

    # Plot the bar chart using 'bar' kind, with specified colors, edgecolor, and linestyle
    df_grouped.plot(kind='bar', ax=ax, color='skyblue', edgecolor='black', linewidth=2, linestyle='-', alpha=0.7)

    # Add a title with a colored background for better visibility
    plt.title(title, bbox={'facecolor': '0.8', 'pad': 3})

    # Label the x-axis with the specified column label
    plt.xlabel(column_label)

    # Label the y-axis with 'Count'
    plt.ylabel('Count')

    # Rotate x-axis labels for better readability
    plt.xticks(rotation=0)

    # Display the bar chart
    plt.show()

# Example usage: Plotting the bar chart for the distribution of 'meal' types
plot_bar_chart_from_column(hotel_booking_df, 'meal', 'Preferred Meal Type')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Create a new column to indicate whether the reserved and assigned room types are the same or different
hotel_booking_df['same_room'] = (hotel_booking_df['reserved_room_type'] == hotel_booking_df['assigned_room_type']).map({True: 'same room', False: 'different room'})

# Group the data by 'is_canceled' and the newly created 'same_room' column and calculate the count
not_same_room_count = hotel_booking_df.groupby(['is_canceled', 'same_room']).size().unstack()

# Calculate the percentage of not getting the same room
not_same_room_count['Percentage'] = (not_same_room_count['different room'] / (not_same_room_count['different room'] + not_same_room_count['same room'])) * 100

# Plot the percentage of not getting the same room vs Booking Cancellation
plt.figure(figsize=(12, 6))
sns.barplot(x=not_same_room_count.index, y=not_same_room_count['Percentage'], hue=not_same_room_count.index, palette='Set2')

# Set the title of the plot
plt.title('Percentage of Not Getting the Same Room vs Booking Cancellation')

# Set the label for the x-axis
plt.xlabel('Booking Cancellation')

# Set the label for the y-axis
plt.ylabel('Percentage of Not Getting the Same Room')

# Display the legend for room type comparison
plt.legend(title='Room Type Comparison')

# Display the plot
plt.show()
```
